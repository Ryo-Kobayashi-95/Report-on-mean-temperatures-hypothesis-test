# Comparison of temperatures between Vancouver and Toronto

## Introduction


The mean daily temperatures of a given station is affected by many different factors such as the latitude of a location, the geography of the location, its proximity and orientation relative to a body of water, its proximity to mountains, the altitude of a location and the dominant wind pattern all together determine the climate of a location. For example, the cities of Vancouver and Toronto have different characteristics that affect their climate in different ways for example, Vancouver is in proximity to the Pacific Ocean and is right next to the Strait of Georgia and it is also between mountains of relatively high altitudes. Toronto on the other hand is located more to the South and although it is also close to a body of water, Lake Ontario, it has a different orientation relative to the body of water, there are also no significant mountains that are in proximity to the city of Toronto and even though Toronto is close to a body of water it is situated much more inland, in other words much farther from the ocean than Vancouver. In this project, we will try to answer the following question:
#### Is the mean daily temperature during the month of January warmer in Vancouver than in Toronto?

To answer the question, we will use climatological data of both cities main airport, from which we use the climatological conditions of the two cities during the month of January between 2016 and 2020 from the Government of Canada past weather and climate section (Climate and Weather Data: https://climate.weather.gc.ca/).


The null hypothesis is that the two cities have the same mean daily temperature in January and the alternative hypothesis is that Vancouver has a higher mean daily temperature than Toronto in January. 

## Methods and Results

Below we will present the data and illustrate the process that we used to obtain the preliminary estimates needed.

```r
library(tidyverse)
library(repr)
library(broom)
library(infer)
options(repr.plot.width = 13)
# For reproducible results
set.seed(12345)
```

```r
# Importing data from the websource and some tidying
get_years <- function(year, station) read.csv(paste0('https://climate.weather.gc.ca/climate_data/bulk_data_e.html?format=csv&stationID=', station, '&Year=', year, '&Month=12&Day=1&time=&timeframe=2&submit=Download%20Data'), stringsAsFactors=FALSE, encoding='UTF-8')

# Station ids obtained directly from gov website search
tor <- do.call(rbind, lapply(c(2016:2020), get_years, station='51459'))
van <- do.call(rbind, lapply(c(2016:2020), get_years, station='51442'))

stopifnot(all.equal(nrow(tor), nrow(van), 365*5 + 1)) #should have every day in the years we chose

temps <- rbind(tor, van)

print('Table 1: Preliminary Data')
head(temps)
```

#### Table 1: Preliminary Data

The only columns we are interested in are the columns for the mean daily temperature and the station name, so we will clean the dataset to focus on those

```r
# We only really care about the mean temperature as the parameter, so select that (along with other columns to aid in visualization)
# We also select Jan only since that has the sharpest difference that we can obtain a good hypothesis on
sample <- temps %>%
    rename(mean_temp = Mean.Temp...C., date = Date.Time, station = Station.Name) %>%
    select(mean_temp, date, station) %>%
    mutate(date = as.Date(date)) %>%
    filter(strftime(date, "%m") == '01') %>%
    mutate(day = strftime(date, "%d"), year = strftime(date, "%y"))
# Calculating the estimate of the population parameter for the daily mean temperature for the two stations for the whole sample
sample_means <- sample %>%
    select(station, mean_temp) %>%
    group_by(station) %>%
    summarize(mean = mean(mean_temp))

print('Table 2: January Mean Daily Temperature for Vancouver and Toronto')
sample_means
```

#### Table 2: January Mean Daily Temperature for Vancouver and Toronto

```r
# Visualizations of the raw data

# Line plot of all years from the two stations
ggplot(sample) +
    geom_line(aes(x=day, y=mean_temp, group = interaction(station, year), color = station)) +
    xlab('Date') +
    ylab('Temperature (Celcius)') +
    ggtitle('Figure 1: Mean daily temperatures from 2016-2020 in Vancouver and Toronto in January') +
    theme(text = element_text(size = 18))

# Boxplot separating all individual element of the sample into quartiles for both samples and show how they overlap or not
ggplot(sample) +
    geom_boxplot(aes(x=station, y=mean_temp, fill=station)) +
    xlab('Station') +
    ylab('Temperature (Celcius)') +
    ggtitle(' Figure 2: Range of mean daily temperatures from 2016-2020 in Vancouver and Toronto in January') +
    theme(text = element_text(size = 18), plot.title=element_text(size=20))
```

***Figure 1*** illustrates the daily mean temperature through the month of January between 2016 and 2020 for both stations. The blue lines represent the daily mean temperature for Vancouver and the red lines represent the mean daily temperature for Toronto. In this graph we can see that the mean daily temperature in Vancouver is generally above the mean daily temperature in Toronto, at some occasions Toronto has higher mean daily temperatures, but we can see a strong apparent trend toward Vancouver having higher mean daily temperatures than Toronto.


***Figure 2*** illustrates the distribution of temperatures for both stations in boxplots, the boxplots represents the distribution of mean daily temperature in quartiles, the colored boxes represent the second and third quartiles, so the middle-range values and the bars at both end of the box represent the more extreme values for mean daily temperature for both stations. In this graph we see an overlap between the fourth quartile for Toronto and the first quartile of Vancouver mean daily temperatures. So only, the lowest mean daily temperatures of Vancouver overlap with the highest mean daily temperatures of Toronto which would again indicate a difference in the mean daily temperature between the two stations. However, the analysis from these two figures is preliminary, so we cannot draw conclusions from them yet.

We will first attempt the **theory-based** method.

```r
# Key statistical values; t-statistic, p-value, and confidence interval

# Filter the sample data by station.
# sample_mean_Toronto only contains data of the station in Toronto.
sample_mean_Toronto <- sample %>%
    filter(station == "TORONTO INTL A")
# sample_mean_Vancouver only contains data of the station in Vancouver.
sample_mean_Vancouver <- sample %>%
    filter(station == "VANCOUVER INTL A")

# Utilize the 't.test' function to generate a comprehensive table
# with key statistical values, such as t-statistic, p-value, and
# confidence interval for alpha = 0.99, based on the 'mean_temp' variable.
# Specify the 'greater' option for the alternative hypothesis that posits
# "Vancouver has a higher mean daily temperature than Toronto in January."
t_test_result <- tidy(
    t.test(x = sample_mean_Vancouver$mean_temp,
           y = sample_mean_Toronto$mean_temp,
           alternative = "greater",
           conf.level = 0.99))

print('Table 3: T-test result for the null hypothesis of equal mean daily temperature for Toronto and Vancouver')
t_test_result
```

#### Table 3: T-test result for the null hypothesis of equal mean daily temperature for Toronto and Vancouver


This table comes from a t-test conducted from the sample and it suggests that there is a difference between the mean daily temperatures of the two stations of $8.124^{\circ}\text{C}$, with Vancouver being warmer. Based on a normal distribution, there is a $95\%$ confidence that the difference between the two cities mean daily temperature is at least $6.837^{\circ}\text{C}$ with Vancouver being warmer and the p.value is about $3.384 \cdot 10^{-36}$, which is well below the established  significant level of $0.01$. So, according to the t-test value we should reject the null hypothesis stating that there is no difference between the two stations mean daily temperature in January.

```r
# Visualization of the 'sample' distribution

# Plot the sample distribution of the mean daily temperatures in Vacouver and Toronto from the sample data.
# The distribution is visually differentiated by colors corresponding to each city, highlighting the unique distribution in each location.
sample %>%
    ggplot() +
    geom_histogram(aes(x = mean_temp, fill= station),binwidth = 2) +
    labs(title = "Figure 3: Sample distribution of the mean daily temperature in Vancounver and Toronto",
         x = "Sample mean temperature") +
    theme(text = element_text(size = 20), plot.title=element_text(size=20))
```

***Figure 3*** is just another visualization that illustrates the difference in the mean daily temperatures for the two stations with Vancouver data in blue and Toronto data in red. Again we can see that Vancouver has warmer mean daily temperatures than Toronto. From this histogram however, we can approximate the mean of the mean daily temperature for both stations around which the bootstrap distribution should be centered for each station respectively.

And now the **bootstrapping** method:

```r
# Bootstrap distribution

# Conduct a bootstrap analysis by generating 10,000 resamples to estimate
# the distribution of the mean daily temperature difference between Vancouver and Toronto.
# The temperature difference is calculated as the value of Vancouver minus Toronto in each resample.
bootstrap_distribution <- sample %>%
    specify(formula = mean_temp ~ station) %>%
    generate(type = "bootstrap", reps = 10000) %>%
    calculate(stat = "diff in means", order = c("VANCOUVER INTL A", "TORONTO INTL A"))

print('Table 4: Bootstrap difference in mean daily temperature for the two stations')
head(bootstrap_distribution)
```

#### Table 4: Bootstrap difference in mean daily temperature for the two stations

***Table 4*** is just a preview of the difference in means between the two stations for the bootstrap samples generated under the column stat and the column replicate is the number of the bootstrap sample generated.

```r
# Confidence Interval

# Compute the observed difference in mean daily temperature between Vancouver and Toronto using the sample data.
# Utilize the 'pivot_wider' function for data wrangling.
obs_diff <-
    sample %>%
    group_by(station) %>%
    summarise(mean = mean(mean_temp)) %>%
    pivot_wider(names_from = station, values_from = mean) %>%
    mutate(diff = `VANCOUVER INTL A` - `TORONTO INTL A`) %>%
    pull(diff)

print(paste("Observed difference:", obs_diff, ". This number represents the mean difference of the daily mean temperature of the two cities, so our bootstrap distribution center should be close to this number."))

# Calculate the condidence interval with 99% confidence level with get_confidence_interval function.
ci <- bootstrap_distribution %>%
    get_confidence_interval(level = 0.99, type = 'percentile')

print('Table 5: 99% Bootstrap Confidence Interval for the Difference in Mean of January Mean Daily Temperature')
ci
```

#### Table 5: 99% Bootstrap Confidence Interval for the Difference in Mean of January Mean Daily Temperature


***Table 5*** represents the upper and lower bound of the $99\%$ bootstrap two-sided confidence interval. In this case, unlike the t-test we are using a two-sided confidence interval because our bootstrap distribution is centered around the mean of the difference and we want to know what value is most likely to be captured by the confidence interval and we do not want to skew our interval on one side. This interval does not capture a value of $0$ which would indicate that there is no difference between the January mean daily temperature for the two stations, but since this value was not captured by the confidence interval, this means that just like with the t-test we should reject the null hypothesis.

```r
# Visualization of the boostrap distribution

# Employ the 'visualize' function from the 'infer' package to create a plot displaying the bootstrap distribution
# and the 'shade_confidence_interval' function to visually indicate the confidence interval on the plot.
bootstrap_distribution_plot <- bootstrap_distribution %>%
    visualize()+
    shade_confidence_interval(endpoints = ci) +
    ggtitle("Figure 4: Bootstrap distribution of the difference in mean daily temperature between Vancouver and Toronto") +
    xlab("Difference in mean daily temperature") +
    theme(text = element_text(size = 20), plot.title=element_text(size=18))
bootstrap_distribution_plot
```

***Figure 4*** is the visualization of the bootstrap difference in means distribution with the $99\%$ confidence interval shaded in blue. From this distribution we can clearly see that $0$ is very far from being captured by the confidence interval as no bootstrap below a difference of $6^{\circ}\text{C}$ were generated out of the $10000$ generated which would indicate that the probability of the two stations having the same mean daily temperature for the month of January is less than $1/10000$.

#### Methods And Results

Our analysis revolves around January mean daily temperatures of Vancouver and Toronto, which we anticipate will differ due to their distinct geographical locations and climate patterns. General agreement is that the temperature gap between these two cities is greatest in January, and based on this, we expect Vancouver to have a higher mean daily average temperature than Toronto during this month. 


As we could see from ***Figure 1 and 2***, from the line graph, it appears that the mean daily temperatures in Vancouver are more often than not above those in Toronto. The boxplot graph displays the total range of mean daily temperatures in the two cities and we can see that there is some overlap if we look at the more extreme quartiles, more specifically the first and the fourth quartile, but there is a stark difference in the median mean daily temperature for the two cities and the two mean average daily temperature for the two cities are different too. However, the question remains is this difference statiscally significant enough to have a sufficient confidence in the claim that Vancouver has higher mean daily temperatures than Toronto in January, which is why we conducted inference analysis of our data to see how probable it is that Vancouver is warmer than Toronto in January.

To evaluate these two, we will conduct difference in the means sample t-test using assumptions from the central limit theorem and also we will use bootstrapping as a way to confirm the findings of the difference in means sample t-test. We will use a $99\%$ confidence interval to answer our question. We are using a strict confidence interval (i.e.: low significance level of $0.01$ or $1\%$) because mean daily temperatures can vary a lot on a daily basis and also from different month of January depending on the general atmospheric pattern that can lead to abnormally warm or cold temperatures which could skew the data. The sample t-test will be performed with the sample data that we retrieved from the web and the bootstrapping distribution will be done with the same data for which we will create $10000$ bootstrap samples of the same size as our original sample.

Based on our analysis, there are some differences that arise from the t-test and the bootstrap distribution as a method to conduct our analysis. First of all, the confidence interval provided by the t-test has no upper bound since we conducted a right sided alternative hypothesis which in our case is not dramatic, but not very useful either. On the other hand the bootstrap distribution does provide both a lower and upper bound to the confidence interval of the difference in mean daily temperature between the two stations, however, since no bootstrap sample captured the value of $0$ and that $10000$ bootstrap samples were generated, this would indicate a p-value between $0$ and $0.0001$ which is a much coarser approximation than with the t-test that states that the p-value is about $3.384 \cdot 10^{-36}$. However, we think it is important to precise that both methods yielded the same result as they both led to a rejection of the null hypothesis, so in this case we do not think that either one is better as they both have equivalent flaws and advantages. If we had to choose however, the bootstrap sample method would probably be slightly better in this case because of its bounded confidence interval, the t-test method leads to an unphysical statement because of its upper limit to infinity which is impossible because infinite temperatures are impossible due to a limited amount of energy through the universe.
